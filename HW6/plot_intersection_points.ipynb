```python
from typing import Callable
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
```

```python
def plotIntersection(x_range: np.ndarray, f: Callable, g: Callable):

    x_max , x_min = int(np.max(x_range)), int(np.min(x_range))

    #   finding intersection points
    res = fsolve(lambda x: f(x) - g(x), x0=[i for i in range(x_min,x_max,2)])
    #   filter relevant
    res = [i for i in res if np.isclose((f(i) - g(i)), 0, atol=0.00001) ]
    # remove duplicates
    points = list(dict.fromkeys(np.round(res, decimals=3)))

    #   plotting
    plt.plot(x_range, f(x_range), '-', label=f.__doc__)
    plt.plot(x_range, g(x_range), '-', label=g.__doc__)
    plt.plot(points, [f(x) for x in points], 'ro')
    plt.xlim([x_min -1, x_max + 1])
    plt.legend(loc='upper left')
    plt.show()

    return points
```

```python
f = lambda x: x**2
f.__doc__ = 'f(x) = x^2'

g = lambda x: x+10
g.__doc__ = 'g(x) = x+10'
print(plotIntersection(np.linspace(-10,10,1000), f, g))
```
