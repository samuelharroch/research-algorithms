```python
# חישוב תקציב נעשה על סכימה של עמודת "הוצאה נטו" עבור כל שורות הרלוונטיות. (ההנחה שגם הכנסות הן חלק מחישוב תקציב ) 
def education_budget(year:int)->int:
    return df[(df['שם רמה 2'] == 'חינוך') & (df['שנה'] == year )]['הוצאה נטו'].sum()
```

```python
education_budget(2019)
```

```python
import numpy as np 

def security_budget_ratio(year:int)->float:
    security_budget = df[(df['שם רמה 1'] == 'בטחון וסדר ציבורי') & (df['שנה'] == year )]['הוצאה נטו'].sum()
    overall_budget = df[df['שנה'] == year ]['הוצאה נטו'].sum()
    return  np.round((security_budget/overall_budget)*100, decimals=2)
```

```python
# by "שם רמה 2"
def largest_budget_year(office:str)->int:
    office_data = df[df['שם רמה 2'] == office][['שנה','הוצאה נטו']]
    return int(office_data.groupby(by=['שנה']).sum().idxmax())
    
```

```python
largest_budget_year("ראש הממשלה")
```

```python
import matplotlib.pyplot as plt 

def summary_budget_year(year:int):
    budget_year = df[df['שנה'] == year][['שם רמה 2','הוצאה נטו']]
    agg_budget_year = budget_year.groupby(by=['שם רמה 2']).sum().sort_values(by=['הוצאה נטו'])
    
    fig = plt.figure(figsize = (22,10))
    
    x = [office[::-1] for office in agg_budget_year.index ][:-3]
    y = list(agg_budget_year['הוצאה נטו'])[:-3]
    
    plt.bar(x , y)
 
    plt.xlabel(str("משרד \ סעיף")[::-1])
    plt.ylabel(str("תקציב שנתי")[::-1])
    plt.title(str(year)+str(" תקציב לשנת ")[::-1] )
    plt.xticks(rotation = 45)
    
    plt.tight_layout()
    plt.show()
    
    
```

```python
summary_budget_year(2020)
```
